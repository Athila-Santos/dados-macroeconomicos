**Inserir parâmetros do modelo e equações de equílibrio para encontrar variáveis endógenas do modelo: Y e i**

```python
# Variáveis exógenas

M = 25
P = 5
G = 20
T = 20

# Parâmetros do modelo

c0 = 15
c1 = 0.8
A = 15
ir = 0.5
m_ = 6
m1 = 0.03
m2 = 0.5

# as taxas de juros real e nominal se igualam. isto é i = r
```

```python
# Cenário base

# Produto de equilíbrio

Y_otimo = (1/(1 - c1 + ir * (m1/m2))) * (c0 - (c1 * T)
+ A - ((ir*m_)/m2) + ((ir*(M/P))/m2) + G)
Y_otimo

# Taxa de juros de equilíbrio

i_otimo = ((m_/m2) + ((m1/m2) * Y_otimo) - ((1/m2)*(M/P)))

# Consumo quando modelo encontra-se em equilíbrio

C = c0 + (c1 * (Y_otimo - T))

# Investimento quando modelo encontra-se em equilíbrio

I = A - (ir * i_otimo)


# item i)

print("Produto de equilíbrio:", round(Y_otimo,2),"\nTaxa de juros de equilíbrio:", round(i_otimo,2),
      "\nConsumo no equilíbrio geral:", round(C,2), "\nInvestimento no equilíbrio geral:", round(I,2))
```

*ii) Cenário de política fiscal expansionista (G = 40)*

*Quais os novos valores de de Y, r, C e I? Quais os efeitos
nessa economia em comparação com o cenário base? Ocorre crownding out no
investimento?*

```python
#Cenário quando G = 40

G = 40

# mudança em Y, i, C e I

Y_otimo = (1/(1 - c1 + ir * (m1/m2))) * (c0 - (c1 * T) + A - ((ir*m_)/m2) + ((ir*(M/P))/m2) + G)

i_otimo = ((m_/m2) + ((m1/m2) * Y_otimo) - ((1/m2)*(M/P)))

C = c0 + (c1 * (Y_otimo - T))

I = A - (ir * i_otimo)


print("Efeito de política fiscal expansionista nos indicadores",
      "\nProduto:", round(Y_otimo,2),"\nTaxa de juros:", round(i_otimo,2),
      "\nConsumo:", round(C,2), "\nInvestimento:", round(I,2))
```
