```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Definindo o tamanho da simulação
n = 40
time = np.arange(1,n,1)
```

```python
#Definindo os Parâmetros
c0 = 100
c1 = 0.8
c2 = 0.3
A = 400
a = 0.8
ρ = 0.9

print(100*np.ones((n,1)))
```

# **Item i**: Estado Estacionário

```python
#Definindo os valores iniciais das variáveis, assmindo que estão no equilíbrio de longo prazo (Estado Estacionário, EE, SS)
Y = np.zeros((n,1))
C = np.zeros((n,1))
Inv = np.zeros((n,1))

#Definindo as variáveis exógenas
T_bar = 100*np.ones((n,1))
G_bar = 100*np.ones((n,1))
sup_prim = np.ones((n,1))
r = 0.10175*np.ones((n,1))

#Os Choques
u_g = np.zeros((n,1)) #Choque de PF
u_t = np.zeros((n,1)) #Choque de PF
u_r = np.zeros((n,1)) #Choque de PM

#Simulando as variáveis no Estado Estacionário (Equilíbrio de Longo Prazo)
for i in range(2,n,1):
  G_bar[i] = G_bar[i] + u_g[i]
  T_bar[i] = T_bar[i] + u_t[i]
  r[i] = r[i] + u_r[i]
  Y[i] = (1/(1 - c1)) * (c0 - (c1 * T_bar[i]) - (c2 * r[i]) + A - a * r[i] + G_bar[i])
  C[i] = c0 + (c1 * (Y[i] - T_bar[i])) - (c2 * r[i])
  Inv[i] = A - (a * r[i])
  sup_prim[i] = (T_bar[i] - G_bar[i])/Y[i]

#Removendo o valor inicial
j = 2
time = np.arange(j,n,1)
Y = Y[j:]
C = C[j:]
Inv = Inv[j:]
r = r[j:]
sup_prim = sup_prim[j:]
G_bar = G_bar[j:]
T_bar = T_bar[j:]
```

```python
#Configurando a estrutura do plot
plt.figure(figsize=(16, 10), dpi=400)

#Representando graficamente as variáveis
plt.subplot(2, 4, 1)
plt.plot(time, Y, linewidth=2)
plt.title('Produto', color='black', fontweight='bold')
plt.xticks()
plt.yticks(rotation='vertical')
plt.gca().spines['top'].set_linewidth(0.8)
plt.gca().spines['right'].set_linewidth(0.8)
plt.gca().spines['bottom'].set_linewidth(0.8)
plt.gca().spines['left'].set_linewidth(0.8)

plt.subplot(2, 4, 2)
plt.plot(time, C, linewidth=2)
plt.title('Consumo', color='black', fontweight='bold')
plt.xticks()
plt.yticks(rotation='vertical')
plt.gca().spines['top'].set_linewidth(0.8)
plt.gca().spines['right'].set_linewidth(0.8)
plt.gca().spines['bottom'].set_linewidth(0.8)
plt.gca().spines['left'].set_linewidth(0.8)

plt.subplot(2, 4, 3)
plt.plot(time, Inv, linewidth=2)
plt.title('Investimento', color='black', fontweight='bold')
plt.xticks()
plt.yticks(rotation='vertical')
plt.gca().spines['top'].set_linewidth(0.8)
plt.gca().spines['right'].set_linewidth(0.8)
plt.gca().spines['bottom'].set_linewidth(0.8)
plt.gca().spines['left'].set_linewidth(0.8)

plt.subplot(2, 4, 4)
plt.plot(time, G_bar, linewidth=2)
plt.title('Gastos', color='black', fontweight='bold')
plt.xticks()
plt.yticks(rotation='vertical')
plt.gca().spines['top'].set_linewidth(0.8)
plt.gca().spines['right'].set_linewidth(0.8)
plt.gca().spines['bottom'].set_linewidth(0.8)
plt.gca().spines['left'].set_linewidth(0.8)

plt.subplot(2, 4, 5)
plt.plot(time, T_bar, linewidth=2)
plt.title('Impostos', color='black', fontweight='bold')
plt.xlabel('Tempo', fontsize=11)
plt.xticks()
plt.yticks(rotation='vertical')
plt.gca().spines['top'].set_linewidth(0.8)
plt.gca().spines['right'].set_linewidth(0.8)
plt.gca().spines['bottom'].set_linewidth(0.8)
plt.gca().spines['left'].set_linewidth(0.8)

plt.subplot(2, 4, 6)
plt.plot(time, sup_prim, linewidth=2)
plt.title('Superávit Primário', color='black', fontweight='bold')
plt.xlabel('Tempo', fontsize=11)
plt.xticks()
plt.yticks(rotation='vertical')
plt.gca().spines['top'].set_linewidth(0.8)
plt.gca().spines['right'].set_linewidth(0.8)
plt.gca().spines['bottom'].set_linewidth(0.8)
plt.gca().spines['left'].set_linewidth(0.8)

plt.subplot(2, 4, 7)
plt.plot(time, r, linewidth=2)
plt.title('Taxa de Juros', color='black', fontweight='bold')
plt.xlabel('Tempo', fontsize=11)
plt.xticks()
plt.yticks(rotation='vertical')
plt.gca().spines['top'].set_linewidth(0.8)
plt.gca().spines['right'].set_linewidth(0.8)
plt.gca().spines['bottom'].set_linewidth(0.8)
plt.gca().spines['left'].set_linewidth(0.8)
```
